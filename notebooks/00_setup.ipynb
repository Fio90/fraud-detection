## Project overview

This project builds a two-layer fraud detection system:

- **Layer 1 — Credit card fraud:** Detects stolen cards at transaction time using ML (XGBoost, Logistic Regression, Isolation Forest)
- **Layer 2 — Order fraud:** Catches fraudulent orders using behavioural signals, even when Layer 1 is bypassed

**Dataset:** [Credit Card Fraud Detection](https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud) — 284,807 transactions, 0.17% fraud rate
